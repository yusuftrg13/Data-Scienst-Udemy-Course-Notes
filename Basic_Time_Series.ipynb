```python
# prompt: apple_data yı datarame çercevesine al

# No need to reinstall pandas_datareader, it's already installed in the previous code execution

import pandas as pd
import yfinance as yf

# Download Apple stock data for the past 5 years (this part is already in the previous code)
apple_data = yf.download("AAPL", start="2018-07-26", end="2024-12-10")

# Create a pandas DataFrame (apple_data is already a DataFrame)
apple_df = pd.DataFrame(apple_data)

# Print the first few rows of the DataFrame to verify
print(apple_df.head())
```

```python
# prompt: Basit zaman serisi ile grafik oluştur

import matplotlib.pyplot as plt

# Plot the closing price of Apple stock
plt.figure(figsize=(12, 6))
plt.plot(apple_df['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Apple Stock Closing Price')
plt.grid(True)
plt.show()
```
