```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
df.head()
```

Bahşiş isimli bir veri setidir resturant içerisinde durumlara göre tip verme özelliklerini tutmuştur.

```python
#Boxplot
sns.boxplot(x=df["total_bill"]);
```

```python
#Kutu Grafik(Box Plot) Çarprazlamalar
sns.boxplot(x=df["day"],y=df["total_bill"],data=df);
```

```python
#Sabah mı Akşam mı? Daha çok para kazanıyoruz
sns.boxplot(x=df['time'],y=df['total_bill'],data=df)
```

```python
#Kişi sayısı kazanç
sns.boxplot(x=df['size'],y=df['total_bill'],data=df)
```

```python
sns.boxplot(x=df['day'],y=df['total_bill'],hue="sex",data=df)
```

```python
sns.catplot(y="total_bill",kind="violin",data=df)
```

Korelasyon Grafikleri

```python
#Scatter Plot
sns.scatterplot(x=df['total_bill'],y=df['tip'],data=df)
```

```python
#Tip ve Kişi sayısını karşılaştıralım
sns.scatterplot(x=df['total_bill'],y=df['tip'],hue=df['size'],data=df)
```
